# Data Cleaning And Processing

We now have three options for dealing with NaN values in the array-

1.Delete the entire rows for the countries having NaN as their value in the row.But this option is not feasible since we have limited data set.

2.Replacing all the NaN values with zero values.This option is feasible but not for the row of France, as we know due to COVID,International Tourism was not permitted in China and UK but it is not valid for France.So we can delete the France Row.

3.Taking Values for France from Google and online sites.Feasible as this is very small dataset and we can manually put in the row and column value.So we proceed with this option.


```python
Tourism.info()
```

```python
A=Tourism.loc[:,'International Tourist Arrivals - (in million) - 2019'].sum()
B=Tourism.loc[:,'International Tourist Arrivals - (in million) - 2020 *'].sum()
C=Tourism.loc[:,'Percentage (%) Share - 2019'].sum()
D=Tourism.loc[:,'Percentage (%) Share - 2020 *'].sum()

print(A,B,C,D)
```

Now since Percentage share in 2019 and 2020 does not add upto 100,we can find the total values by following:

```python
A_total = A*100/57.45
B_total = B*100/51.78
print(A_total,B_total)
```

```python
Tourism.iloc[0,2] = 217.87
Tourism.iloc[0,3] = 117
Tourism.dropna(how="any")
Tourism=Tourism.fillna(0)
Tourism
```

```python
A_new = Tourism.loc[:,'International Tourist Arrivals - (in million) - 2019'].sum()
B_new = Tourism.loc[:,'International Tourist Arrivals - (in million) - 2020 *'].sum()
C_new = Tourism.loc[:,'Percentage (%) Share - 2019'].sum()
D_new = Tourism.loc[:,'Percentage (%) Share - 2020 *'].sum()

print(A_new,B_new,C_new,D_new)
```

Now as the total in the columns of Percentage share 2019 and 2020 doesn't add upto 100 we can insert the new row in our data set
for World Tourism in other countries.

```python
w = A_total - A_new
x = B_total - B_new
y = w/A_total*100
z = x/B_total*100
Tourism.loc[24,:] = [25,'Others',w,x,y,z]
Tourism
```

Now calculating and filling the Percentage Share in France for 2019,2020.

```python
Tourism.iloc[0,4] = 217.87/A_total*100
Tourism.iloc[0,5] = 117/B_total*100
Tourism
```

```python
# Verification of Percentage sum to 100
A_nnew = Tourism.loc[:,'Percentage (%) Share - 2019'].sum()
B_nnew = Tourism.loc[:,'Percentage (%) Share - 2020 *'].sum()
print(A_nnew,B_nnew)
```

Now our data is preprocessed and ready to be plotted.

# Axes Data Defining

```python
Country = Tourism["Country"]
Tourist_Arrivals_19 = Tourism["International Tourist Arrivals - (in million) - 2019"]
Tourist_Arrivals_20 = Tourism["International Tourist Arrivals - (in million) - 2020 *"]
Percentage_Share_19 = Tourism["Percentage (%) Share - 2019"]
Percentage_Share_20 = Tourism["Percentage (%) Share - 2020 *"]
x = list(Country)
y_1 = list(Tourist_Arrivals_19)
y_2 = list(Tourist_Arrivals_20)
y_3 = list(Percentage_Share_19)
y_4 = list(Percentage_Share_20)
```

# Scatter Plots

1.Scatter Plot for Year 2019

```python

plt.figure(figsize=(6,4),dpi=100)
plt.scatter(x,y_1)
plt.xticks(Country, rotation=90)
plt.xlabel('Country')
plt.ylabel('International Tourist Arrivals 2019 in millions')
plt.title('International Tourist Arrivals Country Wise For Year 2019',color ='b')


plt.figure(figsize=(6,4),dpi=100)
plt.scatter(x,y_3)
plt.xticks(Country, rotation=90)
plt.xlabel('Country')
plt.ylabel('Inter. Tourist Arrivals 2019 Percentage Share')
plt.title('International Tourist Arrivals Country Wise For Year 2019 (Percentage Share)',color ='b')

plt.show()
```

Data Story - As we can clearly see from above **$\color{red}{\text{France}}$** had **the most international Tourist visits for year 2019**.As we go from France to below ranked countries we see an nearly exponentially decreasing trend in the international visits which indicates that France is a clear winner when it comes to international visits.India should compare its statistics such as places to visit in India,Tourist Safety Laws,Tourist Tax Laws,Urbanization,Infrastructure etc with countries such as France,Spain,USA to see where they lack in and improve on it to get more international visits and hence generate more revenues from tourism,generate more jobs etc and make its culture known to world.

# Box Plot

Box Plot for the Year 2019 and 2020

```python
y = np.zeros((len(y_1),2))
for i in range(len(y_1)):
    y[i][0] = y_1[i]
    y[i][1] = y_2[i]

y=y.tolist()
x_1 = Tourism['Country'].tolist()
```

```python
plt.figure(figsize=(6,4),dpi=100)
g = sns.boxplot(y)
g.set_xticks(range(25))
g.set_xticklabels(x_1, rotation = 90)
plt.xlabel('Country')
plt.ylabel('International Tourist Arrivals 2019 and 2020 in millions')
plt.title('Box Plot of International Tourist Arrivals Country Wise For Year 2019 and 2020',color ='b')
```

Data Story - Due to **COVID 19** we can see there is lot of variance between the data of Tourism population for the year 2019 and 2020 for all the countries with **France** having the **maximum variance** and **mean** compared to all other countries.We can also see that **India** has the **lesser variance** compared to other countries conveying the information that India is one of the countries that had **one of the least impact** of COVID 19 for the Tourism population compared to other countries.
